```python


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create an empty dictionary to store value counts
value_counts = {}

# Iterate over each column in the dataframe
for column in df.columns:
    # Calculate the value counts for the current column
    counts = df[column].value_counts()
    # Store the value counts in the dictionary
    value_counts[column] = counts

# Create a count chart for each column with hue='class'
for column, counts in value_counts.items():
    # Create a figure and axis
    fig, ax = plt.subplots()
    # Set the title for the chart
    ax.set_title(f'Value Counts for {column}')
    # Plot the count chart with hue='class'
    sns.set_style('whitegrid')
    sns.countplot(data=df, x=column, hue='class', ax=ax)
    # Show the legend
    ax.legend(title='Class')

    # Save the chart as a PNG file
    #plt.savefig(f'{column}_count_chart.png', format='png', dpi=300)

    # Close the figure to release memory
    plt.show()
```

```python
# Whole dataset has categorical column
# apply labelencoder for converting categorical to numerical
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])
```

```python
df.head().T
```

```python
# Reshape the DataFrame using melt
df_div = pd.melt(df, "class", var_name="Characteristics")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(16, 6))

# Create a violin plot with hue='class'
p = sns.violinplot(ax=ax, x="Characteristics", y="value", hue="class", split=True, data=df_div, inner="quartile", palette="Set1")

# Drop the 'class' column from the DataFrame
df_no_class = df.drop(["class"], axis=1)

# Set the x-axis tick labels and rotate them
p.set_xticklabels(rotation=90, labels=list(df_no_class.columns))

# Save the plot as a PNG image
#plt.savefig("violinplot.png", format="png", dpi=500, bbox_inches="tight")

# Show the plot
plt.show()
```

```python
# Create a figure and set the figure size
plt.figure(figsize=(14, 12))
# Create a correlation heatmap using seaborn
sns.heatmap(df.corr(), linewidths=.1, cmap="Purples", annot=True)

# Rotate the y-axis tick labels
plt.yticks(rotation=0)


# Show the heatmap
plt.show()
```
